```python
# pytorch 逻辑回归
# 数学里：
# 线性回归 目标是找到一个线性函数能够拟合数据使得预测值和真实值误差尽可能小
# 其中参数估计就是找到这个线性函数的参数
# 其中参数估计最常用方法就是最小二乘法 即找到使得（所有真实值和预测值的误差平方的和）的最小值的一组参数
```

```python
#把线性回归的输出值实数R区间映射到（0，1）区间的概率值的模型就是逻辑回归
#即 线性回归模型 + sigmoid函数 = 逻辑回归
```

```python
# 概率：已知观测值和参数 求解输出结果
# 似然：已知观测值和结果 求解（估计）参数
#最大似然估计：对模型参数求导算函数取得最大值的点（使得模型输出值与真实值误差最小的参数）
#即找到一个最符合当前观测数据的概率分布
```

```python
#伯努利分布 0-1分布 常见对数损失函数
#标签真实值y_i是 0或1  预测值y_hat即h(x_i)是输出0或1（失败或成功）的概率 
#最大似然算参数 写出似然函数 取对数简化计算 对参数求偏导令=0 解出参数
#此时参数使得模型预测误差最小
# logL = y_i*log(h(x_i)) + (1-y_i)*log(1-h(x_i))
# 要求最大 加负号- 转换求最小 符合损失函数梯度下降法求损失最小找参数的逻辑
```

```python
# 梯度下降法 不是一个算法 是一个方法 结合参数更新最小化损失函数
# 梯度 最大方向导数 方向最大 模长最大
# 在模型参数更新中 是模型参数的更新值
# W_(t+1) = W_(t) - yita*delta(W)  t = 1.....n
# 其中 学习率是梯度下降的步长 确保以已知“最快”的速度找到最小值点 一般在1e-2~1e-3略好 具体根据模型
# 模型中的所有参数都会被计算梯度并梯度下降参数更新
# 学习率越大 梯度下降幅度越大 超过一定值使梯度下降越难收敛 越小幅度越小下降越慢
# 梯度下降快慢由学习率和权重梯度共同决定 
```
